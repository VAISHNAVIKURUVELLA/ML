#Python is Object Oriented Programming

We are sourrounded with objects might be laptop, phone ,dog infront of us.All are these are real world problems
If we want to represent all these in programming paradigm we would need something know as object oriented Programming
Here we need to understand 2 particular concepts: class and objects

Class: Class is a template/blue print of a real world entites. template would comprise of 2 things Properties of the object and Behaviour of the object

Example: Class : Mobile
         Properties : Colour, Cost, Batterylife       
         Behaviour:  Make calls,watch videos, Play Games
         
         
         
Objects: Objects are specific instance of the class

Example : Class: Mobile
          Object: Apple, Motorola, Lenovo

         

```python
  class Phone:
    def make_calls(self):
        print("Making a Phone call")
        
    def Play_game(self):
        print("Playing Games")
```
